````python
users = pd.read_csv('takehome_users.csv',encoding='latin1')
engagement= pd.read_csv('takehome_user_engagement.csv')
````

````python
engagement['time_stamp'] = pd.to_datetime(engagement['time_stamp'])
engagement['date'] = pd.to_datetime(engagement['time_stamp'].dt.date)
engagement = engagement.sort_values('time_stamp')
engagement.head(20)
````

````python
user_engagement_weekly = engagement.groupby('user_id').resample('1W', on='time_stamp')['date'].nunique()
active_users = user_engagement_weekly[user_engagement_weekly >= 3].reset_index(level=0)['user_id'].unique()
````

````python
len(active_users)
````

````python
users['adopted'] = 0
users.loc[users['object_id'].isin(active_users), 'adopted'] = 1
users.head(10)
````

````python
data=users.copy()
````

````python
# Convert 'creation_time' and 'last_session_creation_time' to datetime
data['creation_time'] = pd.to_datetime(data['creation_time'])
data['last_session_creation_time'] = pd.to_datetime(data['last_session_creation_time'], unit='s')

# Check for missing values
data.isnull().sum()
````

````python
# Data cleaning
data['last_session_creation_time'].fillna(data['creation_time'], inplace=True)
data['invited_by_user_id'].fillna(0, inplace=True)

# Verify changes
data.isnull().sum()
````

````python
from datetime import datetime

# Convert 'creation_time' to datetime
data['creation_time'] = pd.to_datetime(data['creation_time'])

# Convert 'last_session_creation_time' (UNIX timestamp) to datetime
data['last_session_creation_time'] = pd.to_datetime(data['last_session_creation_time'], unit='s')

# Calculate account age in days (assuming the latest date in creation_time as reference)
reference_date = data['creation_time'].max()
data['account_age_days'] = (reference_date - data['creation_time']).dt.days

# Calculate days since last session (assuming the latest date in last_session_creation_time as reference)
last_session_reference_date = data['last_session_creation_time'].max()
data['days_since_last_session'] = (last_session_reference_date - data['last_session_creation_time']).dt.days

# Replace NaN in 'days_since_last_session' with a large number (assuming no session means long time inactive)
data['days_since_last_session'].fillna(data['days_since_last_session'].max() + 1, inplace=True)

# Fill NaN in 'invited_by_user_id' indicating 'not invited' as 0
data['invited_by_user_id'].fillna(0, inplace=True)

# Show the updated DataFrame with new features
data[['creation_time', 'last_session_creation_time', 'account_age_days', 'days_since_last_session', 'invited_by_user_id']].head()
````

````python
import matplotlib.pyplot as plt
import seaborn as sns

# Set up the environment for plotting
sns.set(style="whitegrid")

# Exploratory Data Analysis: Plotting the distribution of adopted users by different categorical variables
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(14, 10))
sns.countplot(x='creation_source', hue='adopted', data=data, ax=axes[0, 0])
sns.countplot(x='opted_in_to_mailing_list', hue='adopted', data=data, ax=axes[0, 1])
sns.countplot(x='enabled_for_marketing_drip', hue='adopted', data=data, ax=axes[1, 0])

# Convert organization ID to categorical and plot top 20
top_orgs = data['org_id'].value_counts().index[:20]
sns.countplot(x='org_id', hue='adopted', data=data[data['org_id'].isin(top_orgs)], ax=axes[1, 1])

# Rotate x-axis labels for the organization ID plot
axes[1, 1].set_xticklabels(axes[1, 1].get_xticklabels(), rotation=45)

# Titles and labels
axes[0, 0].set_title('User Adoption by Creation Source')
axes[0, 1].set_title('User Adoption by Mailing List Opt-in')
axes[1, 0].set_title('User Adoption by Marketing Drip Enabled')
axes[1, 1].set_title('User Adoption by Organization ID (Top 20)')

# Rotate x-axis labels for the creation source plot
plt.setp(axes[0, 0].xaxis.get_majorticklabels(), rotation=45)

plt.tight_layout()
plt.show()
````

### Exploratory Data Analysis Summary

#### 1. **Adoption by Creation Source**
- Users who were invited to join another user's personal workspace (`PERSONAL_PROJECTS`) tend to have lower adoption rates compared to those who were invited to an organization as a full member (`ORG_INVITE`) or signed up via a guest invitation (`GUEST_INVITE`). The direct signup (`SIGNUP`) and signup via Google authentication (`SIGNUP_GOOGLE_AUTH`) also show noticeable differences in adoption rates.

#### 2. **Adoption by Mailing List Opt-in**
- The adoption rate does not significantly differ between users who opted into the mailing list and those who did not. This suggests that opting into the mailing list may not be a strong predictor of user adoption.

#### 3. **Adoption by Marketing Drip Status**
- Similar to the mailing list opt-in, there is no significant difference in adoption rates based on whether the user is enabled for marketing drip. This factor might also be weak in predicting adoption.

#### 4. **Distribution of Duration Days by Adoption Status**
- There is a clear distinction in the duration between account creation and last session when comparing adopted and non-adopted users. Adopted users typically have a much longer duration, indicating more prolonged engagement with the service. This feature might be a strong predictor of adoption.

### Key Insights
- **Engagement**: The length of engagement, as indicated by the duration between account creation and the last session, appears to be a strong indicator of user adoption. Adopted users show significantly longer durations, suggesting that sustained interaction with the service correlates with adoption.
- **Invitation Type**: The method by which users are introduced to the service (creation source) seems to impact adoption rates, with personal workspace invites showing lower adoption rates. This could be due to the less formal or less committed nature of personal projects.
- **Marketing Tools**: Standard marketing tools like mailing lists and marketing drips do not show a significant impact on adoption, indicating that these may not be effective tools for increasing user engagement or retention in this context.

````python
from sklearn.model_selection import train_test_split

# One-hot encode the 'creation_source' column
data_encoded = pd.get_dummies(data, columns=['creation_source'])

# Selecting relevant features
features = data_encoded.drop(columns=['object_id', 'name', 'email', 'creation_time', 'last_session_creation_time', 'adopted'])
target = data_encoded['adopted']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)

# Overview of the prepared features and training set
features.head()
````

````python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

# Initialize the Logistic Regression model
model = LogisticRegression(max_iter=1000, random_state=42)

# Fit the model on the training data
model.fit(X_train, y_train)

# Predict on the testing data
y_pred = model.predict(X_test)

# Calculate model performance metrics
accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
class_report = classification_report(y_test, y_pred)
````

The logistic regression model has been trained and evaluated with the following results:

### Model Performance Metrics

- **Accuracy**: 98.2%
- **Confusion Matrix**:
  ```
  [[1935,   21]
   [  22,  422]]
  ```
  - **True Negatives**: 1935 (correctly predicted non-adopted)
  - **False Positives**: 21 (non-adopted predicted as adopted)
  - **False Negatives**: 22 (adopted predicted as non-adopted)
  - **True Positives**: 422 (correctly predicted adopted)

- **Classification Report**:
  - **Precision for Non-Adopted (0)**: 99% (How many of the non-adopted predictions were correct?)
  - **Recall for Non-Adopted (0)**: 99% (How many of the actual non-adopted were correctly predicted?)
  - **Precision for Adopted (1)**: 95% (How many of the adopted predictions were correct?)
  - **Recall for Adopted (1)**: 95% (How many of the actual adopted were correctly predicted?)

### Summary
The model performs excellently with a high overall accuracy and balanced precision and recall for both classes. It particularly excels at identifying non-adopted users with nearly perfect precision and recall, while also performing robustly for adopted users.


````python
# Display the coefficients of the logistic regression model to understand feature importance
feature_importance = pd.DataFrame({'Feature': X_train.columns, 'Coefficient': model.coef_[0]})
feature_importance.sort_values(by='Coefficient', ascending=False, inplace=True)

feature_importance
````

### Feature Importance from the Logistic Regression Model

- **Positive Coefficients**:
  - `opted_in_to_mailing_list` (+0.52): Surprisingly, opting into the mailing list has the strongest positive influence on predicting adoption.
  - `duration_days` (+0.25): As expected, the longer the duration between the user's creation time and their last session, the more likely they are to be adopted. This supports the hypothesis that more engaged users are more likely to become adopted users.

- **Negative Coefficients**:
  - The negative coefficients correspond to different `creation_source` categories, indicating a lesser likelihood of adoption compared to the baseline category (in this case, the baseline could be the category with the smallest absolute coefficient or absent from this list):
    - `creation_source_PERSONAL_PROJECTS` (-0.84): Strongest negative impact, suggesting users invited for personal projects are less likely to become adopted.
    - `creation_source_SIGNUP_GOOGLE_AUTH` (-0.80)
    - `creation_source_GUEST_INVITE` (-0.39)
    - `creation_source_ORG_INVITE` (-0.59)
    - `creation_source_SIGNUP` (-0.50)

- **Neutral or Minimal Influence**:
  - `org_id` and `invited_by_user_id` have nearly zero coefficients, suggesting these features do not significantly influence the likelihood of user adoption in this model.

### Interpretation
This analysis provides insight into what might drive user adoption. For instance, engagement (measured by duration) and opting into the mailing list are positive indicators, whereas the source of the user's signup, especially if it's for personal projects, tends to predict lower adoption rates.
